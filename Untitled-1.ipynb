```python
import math
class Cercle:
    def __init__(self,a,b,r):
        self.a = a
        self.b = b
        self.rayon = r
    def area(self):
        surface = math.pi*self.rayon**2
        print(f"la surface du cercle est {surface}")
    def perimeter(self):
        perimetre = 2*math.pi*self.rayon
        print(f"le perimetre du cercle est {perimetre}")
    def testApprence(self,x,y):
        test = (x-self.a)**2 + (y-self.b)**2-self.rayon**2
        if test == 0:
            print("le point appartient au cercle")
        else:
            print("le point n'appartient par au cercle")

cercle1 = Cercle(3,5,2)
cercle1.area()
cercle1.perimeter()
cercle1.testApprence(8,8)
```

```python
from tkinter import *
import random

root = Tk()

labelJ1 = Label(root, text="J1: 0", font=("", 20), bg="red")
labelJ1.grid(row=5, column=1)
labelJ2 = Label(root, text="Ordi: 0", font=("", 20))
labelJ2.grid(row=5, column=2)
root.geometry("500x600")

buttons = []
for i in range(4):
    l = []
    for j in range(4):
        button = Button(root, text="?", width=3, command=lambda i=i, j=j: retournement(i, j), font=("", 50), fg="red")
        button.grid(row=i, column=j)
        l.append(button)
    buttons.append(l)

joueur1 = True  # True = Joueur humain, False = Ordinateur
j1s = 0
j2s = 0
liste = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'] * 2
random.shuffle(liste)  # Mélange les cartes au début
carteretourner = []


def retournement(lig, col):
   
    global joueur1, j1s, j2s
    index = lig * 4 + col

    if buttons[lig][col]['text'] != "?":  # Empêcher de cliquer sur une carte déjà retournée
        return

    buttons[lig][col].configure(text=liste[index])
    carteretourner.append((lig, col))

    if len(carteretourner) == 2:
        l1, c1 = carteretourner[0]
        l2, c2 = carteretourner[1]

        if liste[l1 * 4 + c1] == liste[l2 * 4 + c2]:  # Si c'est une paire
            if joueur1:
                j1s += 1
                labelJ1.config(text=f"J1: {j1s}")
            else:
                j2s += 1
                labelJ2.config(text=f"Ordi: {j2s}")
                root.after(1000, tour_ordinateur) 
                
        else:
            root.after(500, lambda: refermer(l1, c1, l2, c2))
            joueur_suivant()

        carteretourner.clear()


def refermer(l1, c1, l2, c2):
   
    buttons[l1][c1].config(text="?")
    buttons[l2][c2].config(text="?")


def joueur_suivant():
   
    global joueur1
    joueur1 = not joueur1

    if joueur1:
        labelJ1.config(bg="red")
        labelJ2.config(bg="white")
    else:
        labelJ1.config(bg="white")
        labelJ2.config(bg="red")
        root.after(1000, tour_ordinateur) 

n = []
def tour_ordinateur():
    
    coups_possibles = [(i, j) for i in range(4) for j in range(4) if buttons[i][j]['text'] == "?"]

    if len(coups_possibles) < 2:
        return  

    coup1 = random.choice(coups_possibles)
    retournement(*coup1)
    n.append(coup1)
    
    coups_possibles.remove(coup1)  # Empêcher l'ordinateur de retourner la même carte deux fois
    coup2 = random.choice(coups_possibles)
    n.append(coup2)
    root.after(1000, lambda: retournement(*coup2))  # Retourne la deuxième carte après 1s


root.mainloop()
```

```python
import socket
import sqlite3
import tkinter as tk
from tkinter import ttk
import threading

# --- Fonction du serveur ---
def run_server():
    # Connexion à la base de données
    conn = sqlite3.connect("jeu.db")
    cursor = conn.cursor()

    # Créer la table des messages (si elle n'existe pas déjà)
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS messages (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            message TEXT
        )
    """)
    conn.commit()

    server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server.bind(('localhost', 12345))  # Adresse du serveur et port
    server.listen(1)

    print("Serveur en attente de connexion...")

    # Accepter la connexion du client
    client, address = server.accept()
    print(f"Connexion établie avec {address}")

    while True:
        # Recevoir un message du client
        message = client.recv(1024).decode()
        if message == "exit":
            print("Le client a quitté")
            break

        print(f"Message du client: {message}")

        # Sauvegarder le message dans la base de données
        cursor.execute("INSERT INTO messages (message) VALUES (?)", (message,))
        conn.commit()

        # Envoyer une réponse au client
        client.send("Message reçu et enregistré!".encode())  # Réponse au client

    client.close()
    conn.close()

# --- Fonction du client ---
def envoyer_message():
    message = entry.get()  # Récupérer ce que l'utilisateur a écrit
    client.send(message.encode())  # Envoyer le message au serveur

    # Recevoir la réponse du serveur
    response = client.recv(1024).decode()  # Recevoir la réponse du serveur (taille max: 1024 bytes)
    
    label.config(text=f"Réponse du serveur: {response}")  # Afficher la réponse dans l'interface Tkinter

    # Afficher tous les messages enregistrés dans la base de données
    afficher_messages()

def afficher_messages():
    # Effacer l'affichage actuel
    for row in tree.get_children():
        tree.delete(row)

    # Requête pour obtenir tous les messages de la base de données
    cursor.execute("SELECT * FROM messages")
    rows = cursor.fetchall()

    # Afficher les messages dans l'arbre (tableau)
    for row in rows:
        tree.insert("", tk.END, values=row)

def run_client():
    global client
    # Connexion au serveur
    client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client.connect(('localhost', 12345))

    # Lancer l'interface Tkinter
    root = tk.Tk()
    root.title("Client de Chat")

    # Interface utilisateur
    global entry
    entry = tk.Entry(root, font=("Arial", 14))
    entry.pack(padx=20, pady=10)

    send_button = tk.Button(root, text="Envoyer", font=("Arial", 14), command=envoyer_message)
    send_button.pack(pady=10)

    global label
    label = tk.Label(root, text="", font=("Arial", 14))
    label.pack(pady=10)

    global tree
    columns = ("ID", "Message")
    tree = ttk.Treeview(root, columns=columns, show="headings", height=8)
    tree.heading("ID", text="ID")
    tree.heading("Message", text="Message")
    tree.pack(padx=20, pady=10)

    # Afficher les messages existants dès le début
    afficher_messages()

    root.mainloop()

# --- Exécution ---
if __name__ == "__main__":
    # Créer un thread pour exécuter le serveur
    server_thread = threading.Thread(target=run_server, daemon=True)
    server_thread.start()

    # Exécuter le client
    run_client()
```
